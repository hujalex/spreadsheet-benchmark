```python
transactions = pd.read_csv("data/transactions_data.csv")
```

```python
df_transactions = transactions.head(1000)
```

```python
transactions
```

```python
import pandas as pd

# 1. Remove everything except numbers, dots (decimal), and the negative sign
# This handles cases like '$1,234.56' or '$-77.00'
df_transactions['amount'] = df_transactions['amount'].str.replace(r'[^\d.-]', '', regex=True)
df_transactions['amount'] = pd.to_numeric(df_transactions['amount'])
```

```python
res = df_transactions.groupby('merchant_state', as_index = False)['amount'].mean().nlargest(1, 'amount')
```
